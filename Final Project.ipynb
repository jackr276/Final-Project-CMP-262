## Cleaning the data

As you can see in the dataframes, the salaries are written with dollar signs and commas. This is nice for formatting, but since we'll be using these as numbers in our analysis, it would be best if we cleaned this up

```python
"""
No longer needed, using cell below
"""
def remove_bad_chars(df):
    bad_chars = [",", "$"] #we don't want any of these when we convert to a float
    for index, row in df.iterrows():
        row['starting_salary'] = ''.join(filter(lambda i: i not in bad_chars, row['starting_salary']))
        row['mid_median_salary'] = ''.join(filter(lambda i: i not in bad_chars, row['mid_median_salary']))
        row['percentile_10th'] = ''.join(filter(lambda i: i not in bad_chars, row['percentile_10th']))
        row['percentile_25th'] = ''.join(filter(lambda i: i not in bad_chars, row['percentile_25th']))
        row['percentile_75th'] = ''.join(filter(lambda i: i not in bad_chars, row['percentile_75th']))
        row['percentile_90th'] = ''.join(filter(lambda i: i not in bad_chars, row['percentile_90th']))
```

```python
def clean_currency(x):
    if isinstance(x, str):
        return x.replace("$", "").replace(",","")
    return x


def clean_currency_values(df):
    df.starting_salary = df.starting_salary.apply(clean_currency)
    df.mid_median_salary = df.mid_median_salary.apply(clean_currency)
    df.percentile_10th = df.percentile_10th.apply(clean_currency)
    df.percentile_25th = df.percentile_25th.apply(clean_currency)
    df.percentile_75th = df.percentile_75th.apply(clean_currency)
    df.percentile_90th = df.percentile_90th.apply(clean_currency)
```

```python
clean_currency_values(salaries_by_region_df)
salaries_by_region_df
```

```python
clean_currency_values(salaries_by_college_df)
salaries_by_college_df
```

```python
clean_currency_values(degrees_that_pay_df)
degrees_that_pay_df
```

### Of all colleges/universities in each category(party, polytechnic, etc.), which ones have the lowest/highest salaries overall?

```python
salaries_by_college_df
```

```python
salaries_by_college_df['mid_median_salary'] = pd.to_numeric(salaries_by_college_df['mid_median_salary'])
grouped_by_type_median = salaries_by_college_df.groupby('type').agg({'mid_median_salary':'mean'}).reset_index()
grouped_by_type_min = salaries_by_college_df.groupby('type').agg({'mid_median_salary':'min'}).reset_index()
grouped_by_type_max = salaries_by_college_df.groupby('type').agg({'mid_median_salary':'max'}).reset_index()
```

```python
plot = sns.barplot(data = grouped_by_type_min.sort_values(by = 'mid_median_salary'), x = 'type', y='mid_median_salary', palette=sns.color_palette("rocket")).set_title("Minimum Salary by College Type")
```

```python
plot = sns.barplot(data = grouped_by_type_max.sort_values(by = 'mid_median_salary'), x = 'type', y='mid_median_salary', palette=sns.color_palette("rocket")).set_title("Maximum Salary by College Type")
```

Unlike before, the Ivy League consistently pulls ahead of Engineering colleges here. The median salaries for the Ivy League stay at the top, with Engineering not far behind. Surprisingly, party colleges outrank state colleges in some instances.

### What are the degrees/fields of study that give students the most upward mobility(highest percent-change to mid-career salary change)?

```python
degrees_that_pay_df
```

```python
degrees_that_pay_df['salary_%_change'] = pd.to_numeric(degrees_that_pay_df['salary_%_change'])
sns.barplot(data = degrees_that_pay_df.sort_values(by = 'salary_%_change', ascending=False).head(), x = 'salary_%_change', y = 'major', orient='h', palette=sns.diverging_palette(250, 30, l=65, center="dark")).set_title("Top 5 Highest Percent Increase in Salary by Degree")
```

```python
sns.barplot(data = degrees_that_pay_df.sort_values(by = 'salary_%_change', ascending=False).tail(), x = 'salary_%_change', y = 'major', orient='h', palette=sns.diverging_palette(250, 30, l=65, center="dark")).set_title("Bottom 5 Lowest Percent Increase in Salary by Degree")
```

As we can see here, Math, Philosophy, International Relations, Economics and Marketing degrees give, on average, a doubling in salary from the start to the middle of the degree holder's career. On the other hand, Education, Interior Design, Nutrition, Nursin and P.A.s have the lowest percent increase in salary. It is interesting how no Engineering majors appear here, despite having appeared on the top earners chart from before.
